Array Manipulation
------------------

Remember the `where` function in IDL?  It returned a list of the pixel indices where the input array evaluated to `True`.  For example:

    x = [0,0,1,1]
    print,where(x)
    2 3

`numpy` has a similar function, but it behaves a little differently (and isn't needed as often):

    In [1]: np.where(x)
    Out[1]: (array([2, 3]),)


The return type is a *tuple* that contains an *array*.  Why?  Well, have a look at the 2-d version

    In [1]: x = np.diag([1,2,3])

    In [2]: x
    Out[2]: 
    array([[1, 0, 0],
           [0, 2, 0],
           [0, 0, 3]])
    
    In [3]: np.where(x)
    Out[3]: (array([0, 1, 2]), array([0, 1, 2]))


It returns a `tuple` again, but this time with two arrays.  The arrays are lists of the x and y indices.  This is nice if you want to plot things:

    In [4]: yind,xind = np.where(x)

    In [5]: plot(xind,yind,'o')

We'll be doing something vaguely like this with images.  However, we're probably going to get our x and y coordinates in a different format.  Our lists will look like:

    [(0,0),(1,1),(2,2)]

where each element is an x,y pair, instead of:

    [(0,1,2),(0,1,2)]

where each element is a list of values in one coordinate.

```python
# transform between these
xy = np.array([(0,1,2),(4,3,2)])
print xy
```

```python
# switch to the other way
print xy.T
```

```python
# extract "x" and "y" from xy:
x,y = xy
print "x =",x
print "y =",y
```

One other trick.  You can add arrays elementwise: 

    array([1,2,3]) + array([1,2,3]) = array([2,4,6])

but if you add lists, they just stick to each other:

    [1,2,3]+[1,2,3] = [1,2,3,1,2,3]

similarly, multiplying a list by a constant increases its length:

    [1,2,3]*2 = [1,2,3,1,2,3]

Recursion Revisited
-------------------

Recursion is a pretty challenging concept, and I'm asking you to use it for the final project.  So to help you understand what it's doing, we'll go through a few examples in more detail.

```python
# the simple factorial function
def factorial(x):
    if x == 0:
        return 1
    else:
        return x*(factorial(x-1))    
```

```python
# ignore this cell (at least for now) - it contains some nifty python tricks that are good for teaching,
# but you don't need to know them
def trace(f):
    f.indent = 0
    def g(x):
        print('|  ' * f.indent + '|--', f.__name__, x)
        f.indent += 1
        value = f(x)
        print('|  ' * f.indent + '|--', 'return', repr(value))
        f.indent -= 1
        return value
    return g

def trace2(f):
    f.indent = 0
    def g(x):
        if x > 0:
            print 'factorial( '+repr(x)+' *',
        else:
            print 'factorial( '+repr(x),
        value = f(x)
        print ')',
        return value
    return g
```

```python
# but instead of the actual factorial, let's look at *what it's doing*
# this is a little trick to print out each call of factorial as it happens
@trace2
def factorial(x):
    if x == 0:
        return 1
    else:
        return x*(factorial(x-1))    
factorial(5)
```

```python
# we can also look at the "hierarchy of calls".  It's basically the same thing, but shown differently
@trace
def factorial(x):
    if x == 0:
        return 1
    else:
        return x*(factorial(x-1))    
factorial(5)
```

There are some important points to pay attention to:

1. We do not have a variable tracking the factorial value 
2. Nothing actually "happens" until the last call of `factorial` is written down.   That probably doesn't make a lot of sense, but let's put it another way: the first multiplication that actually happens is `1*factorial(0) = 1*1`.  The *lowest level* happens *first* even though it's set up *last*

```python
# let's do the same with fibonacci
@trace
def fibonacci(x):
    if x == 0:
        return 0
    elif x == 1:
        return 1
    else:
        return fibonacci(x-1)+fibonacci(x-2)
fibonacci(3)
```

This is a good deal more complicated because there are multiple calls at each level.  Try the same for `fibonacci(4)`, `fibonacci(5)`, and maybe even `fibonacci(6)`

What if, instead of the factorial, we want a list of the numbers multiplied to get the factorial?  i.e., instead of 120, we want the numbers (5,4,3,2,1)?  We can do this with essentially the same function, we just replace the *multiplication* operator with a different one (and change the data type)


```python
def factorial_list(x):
    if x == 0:
        return []
    else:
        return [x] + factorial_list(x-1)
```

```python
factorial_list(5)
```

The base case is, "multiply by nothing if x is 0".  We could also have put `[1]` there, since it has the same effect.

`factorial_list` will always return something of type `list`.  So you can always add a `factorial_list` to another (it's just adding two lists).  This is a great way to "track your progress" in a recursive function.

What if we do the same with fibonacci?

```python
def fibonacci(x):
    if x == 0:
        return 0
    elif x == 1:
        return 1
    else:
        return fibonacci(x-1)+fibonacci(x-2)

def fibonacci_list(x):
    if x == 0:
        return [0]
    elif x == 1:
        return [1]
    else:
        return fibonacci_list(x-1)+fibonacci_list(x-2)
```

```python
print "fib(3)=2: ",fibonacci_list(3)
print "fib(4)=3: ",fibonacci_list(4)
print "fib(5)=5: ",fibonacci_list(5)
print "fib(6)=8: ",fibonacci_list(6)
```

What if, instead, we just want to know the *number of function calls*?  This gets to be a little weird and interesting, but still straightforward.

Now, each time, we just need to return a number corresponding to how many times the function is called, *not* the value it should return...

```python
def fibonacci_counts(x):
    if x == 0:
        # no new calls
        return 0
    elif x == 1:
        # no new calls
        return 0
    else:
        # 2 new calls!
        return 2+fibonacci_counts(x-1)+fibonacci_counts(x-2)
```

```python
print "fib(3)=2: ",fibonacci_counts(3)
print "fib(4)=3: ",fibonacci_counts(4)
print "fib(5)=5: ",fibonacci_counts(5)
print "fib(6)=8: ",fibonacci_counts(6)
print "fib(7)=13: ",fibonacci_counts(7)
print "fib(8)=21: ",fibonacci_counts(8)
```

All that matters is that the "form" of the function is the same and that the end conditions are the same.

```python
# an example showing how you could plot the fibonacci number 
# against the number of calls required to calculate it
fibonacci_number = [fibonacci(ii) for ii in xrange(11)]
fibonacci_ncalls = [fibonacci_counts(ii) for ii in xrange(11)]
figure()
plot(fibonacci_number, fibonacci_ncalls, label="Number of calls for each Fibonacci #")
plot(fibonacci_number, fibonacci_number, label="The Fibonacci number vs. itself")
plot(fibonacci_number, np.array(fibonacci_number)*2, label="The Fibonacci number*2 vs. itself")
legend(loc='upper left')
```

Last recursion example: What if we want to visit each pixel in an image, and keep track of the locations that we've visited as a list of coordinates?  We'll start simple:

```python
image = np.array([[1,2],
                  [3,4]])
figure()
imshow(image)
```

Our recursive algorithm will move *right* then *up* (so it must start at the lower left).

We start at [0,0], then our recursive function will move both *right* and *up*, but *right* first.

So we get [0,0], then [1,0] because [1,0] is right of [0,0].  Once we've "moved to" [1,0], the recursive function gets called again!  So it will try to move right from [1,0], but it can't because that's the end.  So then, we move *up* from [1,0]!

This is all a little complicated, but not too terrible to visualize.  And ironically, quite easy to write: most of the code in the following cell is for plotting; all of the meat fits onto 10 lines and could be squeezed onto 8 or 9 pretty easily.

```python
# for display purposes:
import time, sys
from IPython.display import clear_output

def make_arrow(x,y,dx,dy):
    gca().arrow(x,y,dx,dy,head_width=0.1,head_length=0.1) 
    time.sleep(0.5)
    clear_output()
    display(gcf())

    
def traverse_image(x, y, image):
    if x >= image.shape[1]:
        return []
    if y >= image.shape[0]:
        return []
    else:
        gca().plot(x,y,'o') # make circle @ current position
        make_arrow(x,y,1,0) # right
        right = traverse_image(x+1,y,image)  # right
        make_arrow(x,y,0,1) # up
        up = traverse_image(x,y+1,image) # up          
        value_list = ([image[y,x]] + up + right) 
        return value_list
```

```python
# this is a (somewhat choppy) animation of how the navigation happens.  
#Run it a couple times to see what's going on
# NOTE: If you want to see *each step* all plotted in different figures, comment out the "clear_output()" line above
# and re-run this cell
figure()
imshow(image)
axis([-0.5,2.5,-0.5,2.5])
traverse_image(0,0,image)
```

Note that the order of cells visited is 1,3,4,2,4: cell #4, the upper-right cell, is visited twice!  Do you see why?  This is why we need to use the "mask" to mark where we've been in the image photometry recursor.

In keeping with the rest of the tutorial, now ask: What if we want the *coordinates* we've visited instead of the *values at that coordinate* (which is what `[1,3,4,2,4]` is)?

```python
def traverse_image_coordinates(x, y, image):
    if x >= image.shape[1]:
        return []
    if y >= image.shape[0]:
        return []
    else:
        right = traverse_image_coordinates(x+1,y,image)  # right
        up = traverse_image_coordinates(x,y+1,image) # up          
        coord_list = ([(x,y)] + up + right) 
        return coord_list
```

```python
print traverse_image_coordinates(0,0,image)
```

The only change (besides removing the plot commands) is that we changed `image[y,x]`, which got the value at the current position, to `(x,y)`, which is the coordinate pair of the current position.  Keep this in mind as you finish Tutorial 25!
