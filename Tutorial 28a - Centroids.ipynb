One of the functions to write for image registration is an *image centroid* function.

The "centroid" is the middle point inside a star.  It has definitions outside of astronomy, but that's what it means for us.

The centroid is the first "moment" of a function.  Moments are just special sums, defined like this:

$$M_{0} = \sum_{x} \sum_{y} f(x,y)$$
$$M_{x} = \sum_{x} \sum_{y} (x) f(x,y)$$
$$M_{y} = \sum_{x} \sum_{y} (y) f(x,y)$$

Where $f(x,y)$ represents your image (which is, in some sense, "flux as a function of x and y position"), $x$ and $y$ represent the x and y coordinate of each pixel, and $M_{x}$ and $M_{y}$ are the "central moments".  

The centroid, or center pixel of the star is given by:
$$\mu_{x} = \frac{M_{x}}{M_{0}}$$
$$\mu_{y} = \frac{M_{y}}{M_{0}}$$

```python
# we'll start our example by making a 2D grid
xx,yy=np.meshgrid(linspace(-3,3),linspace(-3,3))
```

```python
# they look right
figure()
imshow(xx)
title("X")
colorbar()
figure()
imshow(yy)
title("Y")
colorbar()

# here are some 3D versions so you can see that they're planes in 3D
from mpl_toolkits.mplot3d import axes3d,Axes3D
fig=figure(figsize=[12,6])
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(xx,yy,xx, rstride=1, cstride=1, facecolors=cm.jet((xx-xx.min())/(xx.max()-xx.min())),
        linewidth=0, shade=False, antialiased=False)#, norm=matplotlib.colors.LogNorm())
ax1.view_init(50, -60) 
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(xx,yy,yy, rstride=1, cstride=1, facecolors=cm.jet((yy-yy.min())/(yy.max()-yy.min())),
        linewidth=0, shade=False, antialiased=False)#, norm=matplotlib.colors.LogNorm())
ax2.view_init(50, -60) 
```

```python
# Now we'll make a star
rr = ((xx-2)**2+(yy+1.3)**2)**0.5
gg = (exp(-rr**2/(2*0.5**2)))
imshow(gg)
colorbar()
# 3D again
fig=figure()
ax1 = fig.add_subplot(111, projection='3d')
ax1.plot_surface(xx,yy,gg, rstride=1, cstride=1, facecolors=cm.jet(gg/gg.max()),
        linewidth=0, shade=False, antialiased=False)#, norm=matplotlib.colors.LogNorm())
ax1.view_init(50, -60) 
```

```python
# The sums referenced above are x*f(x,y) and y*f(x,y).  So lets make those images:
fig=figure(figsize=[12,6])
subplot(121)
title("x*f(x,y)")
imshow(xx*gg)
colorbar()
ax1 = fig.add_subplot(122, projection='3d')
ax1.plot_surface(xx,yy,gg*xx, rstride=1, cstride=1, facecolors=cm.jet(gg*xx/(gg*xx).max()),
        linewidth=0, shade=False, antialiased=False)#, norm=matplotlib.colors.LogNorm())
ax1.view_init(50, -60) 
fig=figure(figsize=[12,6])
subplot(121)
title("y*f(x,y)")
imshow(yy*gg)
colorbar()
ax1 = fig.add_subplot(122, projection='3d')
ax1.plot_surface(xx,yy,gg*yy, rstride=1, cstride=1, facecolors=cm.jet(gg/(gg).max()),
        linewidth=0, shade=False, antialiased=False)#, norm=matplotlib.colors.LogNorm())
ax1.view_init(50, -60) 
```

```python
# this is a cut through the image, to give you an idea of what this looks like for something simpler than a star
plot(xx[10,:],'b')
plot(gg[10,:],'g')
plot(xx[10,:]*gg[10,:],'r')
```

```python
# finally, print the X and Y centers
# these are M_x and M_y as defined above
print (xx*gg).sum()/gg.sum(),(yy*gg).sum()/gg.sum()
```
