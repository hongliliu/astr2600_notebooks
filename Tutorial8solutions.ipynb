Finally, some cool trickery with internet access

````python
import urllib
import StringIO
import PIL
f = urllib2.urlopen("http://apod.nasa.gov/apod/image/1301/OrionGEMS_FINAL.jpg").read()
img = PIL.Image.open(StringIO.StringIO(f))
imshow(img)                    
````

You can turn the image into an array:

````python
imgarr = np.array(img)
imgarr.shape
````

There are 3 color "planes" in the image, RGB.  You can show each of them individually:

````python
imshow(imgarr[:,:,0])
````

Do the same thing with the other layers:

````python
imshow(imgarr[:,:,1])
````

````python
imshow(imgarr[:,:,2])
````

````python
hist(imgarr[:,:,0].flat, bins=50, alpha=0.3, color='r')
hist(imgarr[:,:,1].flat, bins=50, alpha=0.3, color='g')
hist(imgarr[:,:,2].flat, bins=50, alpha=0.3, color='b')
````
