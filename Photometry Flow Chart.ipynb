```python
# import the support module
from starphottests import make_image
image = make_image(0)
noise = np.random.randn(*image.shape) * 0.02
noisy_image = image+noise
where_star = noisy_image>0.06
figure(figsize=(12,8),dpi=200)
subplot(121) # subplots are neat; look at the documentation if you want to use them
imshow(noisy_image)
gca().add_artist(Rectangle((72,88),12,14,facecolor='none',edgecolor='w'))
subplot(122)
imshow(where_star)
gca().add_patch(Rectangle((72,88),12,14,facecolor='none',edgecolor='w'))
```

```python
sub_image = noisy_image[88:102,72:84]
sub_mask = where_star[88:102,72:84]#49:59,14:24]
subplot(121)
imshow(sub_image)
subplot(122)
imshow(sub_mask)
```
