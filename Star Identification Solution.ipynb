OK, so how do we select pixels above a certain value?  With the `>` operator.

```python
where_star = noisy_image>0.06
figure(figsize=(12,8),dpi=200)
subplot(121) # subplots are neat; look at the documentation if you want to use them
imshow(noisy_image)
subplot(122)
imshow(where_star)
```

The red regions in the right image indicate the pixels with stars.  The blue regions are "just noise".  

So we now have an "image mask" that tells us where the stars are "statistically significant".  But all we know is whether the pixels are or are not significant, what we actually want is the total flux (total amount of light) for each star.

We already wrote a `get_brightest_chunk` code, so we'll use that to extract a particular star.  Though in the example, because I'm not going to give away the answer for `get_brightest_chunk`, I'll do it by hand.

```python
sub_image = noisy_image[88:102,72:84]
sub_mask = where_star[88:102,72:84]#49:59,14:24]
subplot(121)
imshow(sub_image)
subplot(122)
imshow(sub_mask)
```

We can tell that almost all of the red pixels come from one star.  But not all of them.  So how do we tell which pixels are from one star and which from the other?  We can find a single pixel in one of the stars, then find all of its neighbors that are also "significant", and mark them as part of the same star.  If we find a boundary for our star, we know it is just one star.

One of the best ways to do this is via recursion.  This example is covered in the textbook for IDL, but the array manipulation techniques are a little different in IDL and python so I'll cover the python approach here.

```python
import time, sys
from IPython.display import clear_output

def get_star_flux(x, y, image, mask, draw_arrows=True, fig=None):
    """
    Starting with a point inside a star, find all associated points

    Parameters
    ----------
    x,y : int,int
        The x and y positions of the pixel to add
    image : 2D array
        The image whose fluxes we want to sum
    mask : 2D boolean array
        The mask array telling which pixels are significant
    """
    # Need to check each pixel, and want to check each pixel only once
    # Will set the mask value to zero after checking that pixel
    # Then will add to the flux value...
    if mask[y,x] == 0:
        # end condition: if current pixel is zero, done!
        if draw_arrows:
            f.gca().plot(x,y,marker='o',markerfacecolor='none',markeredgecolor='g')
            time.sleep(0.1)
            #clear_output()
            display(f)
        return 0 
    else:
        if draw_arrows:
            f.gca().arrow(x,y,-1,0,head_width=0.1,head_length=0.1)
            f.gca().arrow(x,y,0,1,head_width=0.1,head_length=0.1)
            f.gca().arrow(x,y,1,0,head_width=0.1,head_length=0.1)
            f.gca().arrow(x,y,0,-1,head_width=0.1,head_length=0.1)
            time.sleep(0.1)
            #clear_output()
            display(f)
        mask[y,x] = 0 # zero the mask (we've now counted this pixel)
        flux = (image[y,x] + # include flux of current pixel
                get_star_flux(x-1,y,image,mask)+ # left
                get_star_flux(x,y+1,image,mask)+ # up
                get_star_flux(x+1,y,image,mask)+ # right
                get_star_flux(x,y-1,image,mask)) # down
        return flux
```
